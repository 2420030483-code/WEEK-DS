```python
import pandas as pd
import numpy as np

print("="*50)
print("DATA CLEANING TECHNIQUES")
print("="*50)

# Sample Dataset
df = pd.DataFrame({
    'Age':[25,30,np.nan,40,35],
    'Department':['HR','Finance','Finance',np.nan,'IT']
})

print("\nOriginal Data")
print(df)

# Mean Imputation
df['Age'].fillna(df['Age'].mean(),inplace=True)

# Mode Imputation
df['Department'].fillna(df['Department'].mode()[0],inplace=True)

print("\nAfter Mean/Mode Imputation")
print(df)

# Forward Fill
df_ffill = df.fillna(method='ffill')

print("\nForward Fill")
print(df_ffill)

# Backward Fill
df_bfill = df.fillna(method='bfill')

print("\nBackward Fill")
print(df_bfill)

# Duplicate Removal
df2 = pd.DataFrame({
    'ID':[1,2,2,3,4,4],
    'Name':['Alice','Bob','Bob','Charlie','David','David']
})

print("\nDataset with Duplicates")
print(df2)

print("\nAfter Removing Duplicates")
print(df2.drop_duplicates())

# Date Standardization
date_df = pd.DataFrame({
    'Date':['2025-01-05','05/01/2025','Jan 5, 2025']
})

date_df['Date'] = pd.to_datetime(
    date_df['Date'],
    errors='coerce'
).dt.strftime('%Y-%m-%d')

print("\nStandardized Dates")
print(date_df)

# Case Normalization
name_df = pd.DataFrame({
    'Name':['Alice','BOB','charlie','DAVID']
})

name_df['Lower'] = name_df['Name'].str.lower()
name_df['Upper'] = name_df['Name'].str.upper()

print("\nCase Normalization")
print(name_df)
```
